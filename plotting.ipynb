```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sals = np.abs(np.load("sals.npy"))
data = np.load("data.npy")
starts = np.load("starts.npy")

```

```python
figs = 5

fig, axs = plt.subplots(figs, 2, figsize=(6, 12))

for i in range(figs):
    axs[i,0].imshow(data[i].T)
    axs[i,0].vlines([starts[i], starts[i]+10], 0, 50, colors='red')
    axs[i,1].imshow(sals[i].T, cmap='Reds')
    axs[i,1].vlines([starts[i], starts[i]+10], 0, 50, colors='red')

plt.tight_layout()
plt.show()
```
